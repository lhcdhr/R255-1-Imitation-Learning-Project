```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
import matplotlib.font_manager as fm


```

# do the plots

```python
def plot_compare_two(m1_random_mean, m1_random_std, m1_expert_mean, m1_expert_std, m1_mean, m1_std, m1_label,
                     m2_random_mean, m2_random_std, m2_expert_mean, m2_expert_std, m2_mean, m2_std, m2_label, title):
    num_trajs= [1,4,10]
    m1_normalised_mean = (m1_mean - m1_random_mean)/(m1_expert_mean-m1_random_mean)
    m1_normalised_std = (m1_std - m1_random_std)/(m1_expert_mean-m1_random_std)
    m2_normalised_mean = (m2_mean - m2_random_mean)/(m2_expert_mean-m2_random_mean)
    m2_normalised_std = (m2_std - m2_random_std)/(m2_expert_mean-m2_random_std)

    plt.plot(num_trajs, m1_normalised_mean,"r", marker='D', markersize=5, label=m1_label)
    plt.fill_between(num_trajs, m1_normalised_mean, m1_normalised_mean-m1_normalised_std, facecolor = 'red', alpha=0.2)
    plt.fill_between(num_trajs, m1_normalised_mean, m1_normalised_mean+m1_normalised_std, facecolor = 'red', alpha=0.2)
    plt.plot(num_trajs, m2_normalised_mean,"b", marker='v', markersize=5, label=m2_label)
    plt.fill_between(num_trajs, m2_normalised_mean, m2_normalised_mean-m2_normalised_std, facecolor = 'blue', alpha=0.2)
    plt.fill_between(num_trajs, m2_normalised_mean, m2_normalised_mean+m2_normalised_std, facecolor = 'blue', alpha=0.2)
    plt.plot(num_trajs, np.ones(3)*1, linestyle="--",color="k", label = "Expert")
    plt.plot(num_trajs, np.zeros(3), linestyle="--",color="g", label = "Random")
    tnr = fm.FontProperties(fname='/content/times.ttf')
    plt.legend(loc="lower right", prop = tnr)
    plt.xticks(num_trajs, fontproperties= tnr)
    plt.yticks(fontproperties = tnr)
    plt.xlabel("Number of Trajectories", fontproperties = tnr)
    plt.ylabel("Normalised Reward", fontproperties = tnr)
    plt.title(title, fontproperties = tnr)
    plt.show()
```

```python
def plot_compare_three(m1_random_mean, m1_random_std, m1_expert_mean, m1_expert_std, m1_mean, m1_std, m1_label,
                       m2_random_mean, m2_random_std, m2_expert_mean, m2_expert_std, m2_mean, m2_std, m2_label,
                       m3_random_mean, m3_random_std, m3_expert_mean, m3_expert_std, m3_mean, m3_std, m3_label, title):

    num_trajs= [1,4,10]
    m1_normalised_mean = (m1_mean - m1_random_mean)/(m1_expert_mean-m1_random_mean)
    m1_normalised_std = (m1_std - m1_random_std)/(m1_expert_mean-m1_random_std)
    m2_normalised_mean = (m2_mean - m2_random_mean)/(m2_expert_mean-m2_random_mean)
    m2_normalised_std = (m2_std - m2_random_std)/(m2_expert_mean-m2_random_std)
    m3_normalised_mean = (m3_mean - m3_random_mean)/(m3_expert_mean-m3_random_mean)
    m3_normalised_std = (m3_std - m3_random_std)/(m3_expert_mean-m3_random_std)

    plt.plot(num_trajs, m1_normalised_mean,"r", marker='D', markersize=5, label=m1_label)
    plt.fill_between(num_trajs, m1_normalised_mean, m1_normalised_mean-m1_normalised_std, facecolor = 'red', alpha=0.2)
    plt.fill_between(num_trajs, m1_normalised_mean, m1_normalised_mean+m1_normalised_std, facecolor = 'red', alpha=0.2)
    plt.plot(num_trajs, m2_normalised_mean,"b", marker='v', markersize=5, label=m2_label)
    plt.fill_between(num_trajs, m2_normalised_mean, m2_normalised_mean-m2_normalised_std, facecolor = 'blue', alpha=0.2)
    plt.fill_between(num_trajs, m2_normalised_mean, m2_normalised_mean+m2_normalised_std, facecolor = 'blue', alpha=0.2)
    plt.plot(num_trajs, m3_normalised_mean,"c", marker='o', markersize=5, label=m3_label)
    plt.fill_between(num_trajs, m3_normalised_mean, m3_normalised_mean-m3_normalised_std, facecolor = 'cyan', alpha=0.2)
    plt.fill_between(num_trajs, m3_normalised_mean, m3_normalised_mean+m3_normalised_std, facecolor = 'cyan', alpha=0.2)

    plt.plot(num_trajs, np.ones(3)*1, linestyle="--",color="k", label = "Expert")
    plt.plot(num_trajs, np.zeros(3), linestyle="--",color="g", label = "Random")
    tnr = fm.FontProperties(fname='/content/times.ttf')
    plt.legend(loc="lower right", prop = tnr)
    #plt.xlim((0,91))
    plt.xticks(num_trajs, fontproperties= tnr)
    plt.yticks(fontproperties = tnr)
    plt.xlabel("Number of Trajectories", fontproperties = tnr)
    plt.ylabel("Normalised Reward", fontproperties = tnr)
    plt.title(title, fontproperties = tnr)
    plt.show()
```
